```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Creating some dummy, noised, S-shaped points:

```python
x = np.linspace(-5, 5, 10)
x += (np.random.random(10)-0.5)/1.5
y = 1 / (1 + np.exp(-x))
y += (np.random.random(10)-0.5)/6
x = np.clip((x+5)/10, 0.01, 0.99)
y = np.clip(y, 0.01, 0.99)
```

Fitting the regresssion model:

```python
plt.figure(figsize=(12, 10))
plt.subplots_adjust(hspace=.5)
plt.subplot(2,2,1)
plt.title('1. Start with some points forming an "S" shape\n\n', fontsize=12)
plt.scatter(x, y)

plt.subplot(2,2,2)
y_logit = np.log(y / (1 - y))
regressor = LinearRegression().fit(x.reshape(-1, 1), y_logit.reshape(-1, 1))
y_predicted = regressor.predict(x.reshape(-1, 1))
plt.title('2. Transforming the points into logits with $\\ln{\\left(\\frac{x}{1-x}\\right)}$\nyields a more or less straight line.\nA linear regression model can be fit to this data.', fontsize=12)
plt.scatter(x, y_logit)
plt.plot(x, y_predicted, color='red')

plt.subplot(2,2,3)
plt.title('3. The regression line can be transformed back to\nthe original domain using sigmoid function $\\frac{1}{1+e^{-x}}$', fontsize=12)
x_new = np.linspace(0,1,100)
y_new = regressor.predict(x_new.reshape(-1, 1))
y_new_sigmoid = 1 / (1 + np.exp(-y_new))
plt.scatter(x, y)
plt.plot(x_new, y_new_sigmoid, color='red')
plt.savefig('sigmoid-regression.svg', bbox_inches='tight')
plt.show()
```
